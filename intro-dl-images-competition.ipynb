# Transfer Learning exercise

Here the goal is that you try yourself to build a binary classifier for a new dataset. The goal is to classify petri-dishes based on the fact that the bacteria in the petri plate did grow or not. The data is coming from a research paper [MicrobIA Images Dataset (Beta ver. 0.1) ](http://www.microbia.org/index.php/resources). This performance will be assessed based on the [cohen kappa score](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.cohen_kappa_score.html) on the test set predictions. You have until the end of the day to try some models and the one that at 5pm has the highest score wins !
